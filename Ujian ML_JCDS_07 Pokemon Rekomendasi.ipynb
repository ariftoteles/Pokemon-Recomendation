Kita akan menggunakan rekomendasi Type1, Generation, dan Legendary, jadi null value pada type 2 kita abaikan

```python
df['Legendary'] = df['Legendary'].map({False:'Not Legend',True:'Legend'})
df['tipe'] = df.apply(lambda i: str(i['Type 1'])+ ' '+ str(i['Generation'])+ ' '+ str(i['Legendary']), axis=1)
df.head(40)
```

```python
cv = CountVectorizer()
cm = cv.fit_transform(df['tipe'])
cosScore = cosine_similarity(cm.toarray())
cosScore
```
